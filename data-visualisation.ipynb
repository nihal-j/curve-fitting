```python
import numpy as np
import pandas as pd
```

```python
# reading the data into a DataFrame
dataDf = pd.read_csv('data.csv')
dataDf.describe()
```

```python
# for interactive notebook plots
%matplotlib notebook

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

# plotting 1000 random points from the dataset

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax1 = plt.axes(projection = '3d')
sample = dataDf.sample(n = 1000)
xdata = sample['LATITUDE'].tolist()
ydata = sample['LONGITUDE'].tolist()
zdata = sample['ALTITUDE'].tolist()
ax1.scatter3D(xdata, ydata, zdata, c=zdata)
```
